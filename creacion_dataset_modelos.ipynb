```python
import pandas as pd
import io
from google.colab import files
import numpy as np
import itertools
import seaborn as sns
from datetime import timedelta
```

```python
def upload_files ():
  uploaded = files.upload()
  for fn in uploaded.keys():
    print('User uploaded file "{name}" with length {length} bytes'.format(
        name=fn, length=len(uploaded[fn])))
    df = pd.read_csv(io.StringIO(uploaded[fn].decode('utf-8')))
    return df
```

#### A la hora de cargar los datos asegurarse que se trabaja con aquellos que hacen referencia a los años **2014-2018**.

```python
games = upload_files()
jugadores = upload_files()
equipos = upload_files()
```

```python
games.GAME_DATE_EST = pd.to_datetime(games['GAME_DATE_EST'], format="%Y/%m/%d")
```

```python
jugadores.columns = ['GAME_ID','TEAM_ID','TEAM_ABBREVIATION', 'TEAM_CITY','PLAYER_ID','PLAYER_NAME','POSICION_INICIO','COMMENT','MINUTOS_JUGADOS','TIROS','TIROS_INTENATDOS','PORCENTAJE_ACIERTO_TIROS',
                          'TRIPLES','TRIPLES_INTENTADOS','PORCENTAJE_TRIPLES','TIROS_LIBRES','TIROS_LIBRES_INTENTADOS','PORCENTAJE_TIROS_LIBRES','REBOTES_OFENSIVOS','REBOTES_DEFENSIVOS',
                          'REBOTES','ASISTENCIAS','ROBOS','TAPONES','ROTACIONES','FALTAS','PUNTOS_ANOTADOS','VALORACION']
```

```python
temporadas = [2014, 2015, 2016, 2017, 2018] # 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013,
```

```python
for i in temporadas:
  x = games[games.SEASON == i].sort_values(['GAME_ID','GAME_DATE_EST']).reset_index()
  
  for j in range(0, len(x)):
    fp = x.GAME_DATE_EST[j]
    id_l = x.HOME_TEAM_ID[j]; id_v = x.VISITOR_TEAM_ID[j]

    #Número de partidos jugados por cada uno de los equipos en los ultimos 7 días
    pj = x[(x.GAME_DATE_EST < fp) & (x.GAME_DATE_EST >= (fp - timedelta(days=7)))]
    ult7_l = len(pj[(pj.HOME_TEAM_ID ==id_l) | (pj.VISITOR_TEAM_ID == id_l)])
    ult7_v = len(pj[(pj.HOME_TEAM_ID ==id_v) | (pj.VISITOR_TEAM_ID == id_v)])

    #Ultimos 5 partidos de cada uno de los equipos
    pj5 = x[(x.GAME_DATE_EST < fp)]
    pj_l = pj5[(pj5.HOME_TEAM_ID == id_l) | (pj5.VISITOR_TEAM_ID == id_l)].reset_index()
    pj_l_5p = pj_l.iloc[(len(pj_l)-5):len(pj_l),:]
    pj_v = pj5[(pj5.HOME_TEAM_ID == id_v) | (pj5.VISITOR_TEAM_ID == id_v)].reset_index()
    pj_v_5p = pj_v.iloc[(len(pj_v)-5):len(pj_v),:]

    #Equipo local
    local_win_l = pj_l_5p[(pj_l_5p.HOME_TEAM_ID == id_l) & (pj_l_5p.HOME_TEAM_WINS == 1)]
    visit_win_l = pj_l_5p[(pj_l_5p.VISITOR_TEAM_ID == id_l) & (pj_l_5p.HOME_TEAM_WINS == 0)]
    winl_5match = len(local_win_l)+len(visit_win_l)

    #Equipo visitante
    local_win_v = pj_v_5p[(pj_v_5p.HOME_TEAM_ID == id_v) & (pj_v_5p.HOME_TEAM_WINS == 1)]
    visit_win_v = pj_v_5p[(pj_v_5p.VISITOR_TEAM_ID == id_v) & (pj_v_5p.HOME_TEAM_WINS == 0)]
    winv_5match = len(local_win_v)+len(visit_win_v)

    datos.loc[datos.GAME_ID == x.GAME_ID[j], "ult7dias_local"] = ult7_l
    datos.loc[datos.GAME_ID == x.GAME_ID[j], "ult7dias_visitante"] = ult7_v
    datos.loc[datos.GAME_ID == x.GAME_ID[j], "win_ult5_local"] = winl_5match
    datos.loc[datos.GAME_ID == x.GAME_ID[j], "win_ult5_visitante"] = winv_5match

    ## ESTADISTICAS 
    ##### Estadisticas Local
    #corte = pj[(pj.HOME_TEAM_ID == id_l)| (pj.HOME_TEAM_ID == id_v)| (pj.VISITOR_TEAM_ID == id_l) | (pj.VISITOR_TEAM_ID == id_v)]
    if len(pj_l_5p) > 0:
      id_games_l = pj_l_5p.GAME_ID
      filter_l = jugadores[('GAME_ID')].isin(id_games_l)
      g_j_l = jugadores[filter_l]
      g_j_l = g_j_l[g_j_l.TEAM_ID == id_l]
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "mean_puntos_local"] = sum(g_j_l['PUNTOS_ANOTADOS'])/len(id_games_l)
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "porcentaje_triples_local"] = sum(g_j_l['TRIPLES'])/sum(g_j_l['TRIPLES_INTENTADOS'])
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "porcentaje_tiros_libres_local"] = sum(g_j_l['TIROS_LIBRES'])/sum(g_j_l['TIROS_LIBRES_INTENTADOS'])
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "mean_rebotes_local"] = sum(g_j_l['REBOTES'])/len(id_games_l)
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "mean_asistencias_local"] = sum(g_j_l['ASISTENCIAS'])/len(id_games_l)
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "mean_robos_local"] = sum(g_j_l['ROBOS'])/len(id_games_l)
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "mean_tapones_local"] = sum(g_j_l['TAPONES'])/len(id_games_l)
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "mean_faltas_local"] = sum(g_j_l['FALTAS'])/len(id_games_l)
      match_player_l = pd.DataFrame(g_j_l.GAME_ID.value_counts())
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "mean_player_local"] = round(match_player_l['GAME_ID'].mean())
    ##### Estadisticas Visitante
    if len(pj_v_5p) > 0:
      id_games_v = pj_v_5p.GAME_ID
      filter_v = jugadores[('GAME_ID')].isin(id_games_v)
      g_j_v = jugadores[filter_v]
      g_j_v = g_j_v[g_j_v.TEAM_ID == id_v]
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "mean_puntos_visitante"] = sum(g_j_v['PUNTOS_ANOTADOS'])/len(id_games_v)
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "porcentaje_triples_visitante"] = sum(g_j_v['TRIPLES'])/sum(g_j_v['TRIPLES_INTENTADOS'])
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "porcentaje_tiros_libres_visitante"] = sum(g_j_v['TIROS_LIBRES'])/sum(g_j_v['TIROS_LIBRES_INTENTADOS'])
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "mean_rebotes_visitante"] = sum(g_j_v['REBOTES'])/len(id_games_v)
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "mean_asistencias_visitante"] = sum(g_j_v['ASISTENCIAS'])/len(id_games_v)
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "mean_robos_visitante"] = sum(g_j_v['ROBOS'])/len(id_games_v)
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "mean_tapones_visitante"] = sum(g_j_v['TAPONES'])/len(id_games_v)
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "mean_faltas_visitante"] = sum(g_j_v['FALTAS'])/len(id_games_v)
      match_player_v = pd.DataFrame(g_j_v.GAME_ID.value_counts())
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "mean_player_visitane"] = round(match_player_v['GAME_ID'].mean())

    #### Numero de victorias conseguidas durante la temporada
    ind_local = victorias.index[victorias["TEAM_ID"] == id_l].tolist()
    ind_visitante = victorias.index[victorias["TEAM_ID"] == id_v].tolist()
    resultado = x.HOME_TEAM_WINS[j]
    if resultado == 1:
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "numero_victorias_local"] = victorias.loc[ind_local,str(i)].values[0]
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "numero_victorias_visitante"] = victorias.loc[ind_visitante,str(i)].values[0]
      victorias.loc[ind_local,str(i)]  = victorias.loc[ind_local,str(i)].values[0] + 1
    else:
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "numero_victorias_visitante"] = victorias.loc[ind_visitante,str(i)].values[0]
      datos.loc[datos.GAME_ID == x.GAME_ID[j], "numero_victorias_local"] = victorias.loc[ind_local,str(i)].values[0]
      victorias.loc[ind_visitante,str(i)]  = victorias.loc[ind_visitante,str(i)].values[0] + 1



```
